```python
def cube(y):
    return y*y*y

lambda_cube=lambda y:y*y*y

# using function defined
#using def key word
print("using function defined with 'def keyword,cube:",cube(5))

#using the lambda function
print("using lambda function ,cube :",lambda_cube(2))
```
